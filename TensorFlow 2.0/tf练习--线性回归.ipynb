```python
import pandas as pd
import numpy as np
import random 
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
a=[]
b=[]
for i in range(0,1001):
    a.append(i)
    b.append(i*3+3.0+random.random()*5)
```

```python
plot.scatter(a,b)
plot.show()
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model_train_log=model.fit(a,b,epochs=1000)
```

```python
predict_weight=model.weights[0]
predict_bias=model.weights[1]
print(predict_weight)
print(predict_bias)
```
