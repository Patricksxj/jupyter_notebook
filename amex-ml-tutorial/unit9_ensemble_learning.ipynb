<strong style="font-size: 125%">第九单元：集成学习</strong>
<br><br>

集成学习是机器学习里重要的方法论。往往最有效的机器学习算法或多或少都有着集成学习的影子。譬如在决策树单元介绍的随机森林算法就是集成学习的一个很好的例子。集成学习的目的是构建多个学习算法，然后通过一定策略综合这些学习算法来完成学习任务，常常可以显著改善单一学习法的效率。

在第四单元里我们介绍过模型复杂度和模型训练之间的关系。下面是对模型复杂度的另外一种诠释。
<br><br>
<div>
    
</div>
<br>

一个很复杂的模型对于训练样本拟合的很好，但是在其它的数据上表现很差，这称为小偏差大方差。一个简单的模型可能在训练样本上拟合的一般，但是在其它的数据上表现的和在训练样本上的指标相类似，这称为大偏差小方差。如果我们在单一的模型基础上能改善其中一个指标，那么最后的模型指标可以有显著的改善。

集成学习法就是利用多个学习算法来改善模型的偏差或方差，达到最终提高学习质量的目的。集成学习法分成三大类：Bagging, Boosting和Stacking。

<strong>Bagging</strong>
<br>

Bagging就是从训练样本中随机取样，然后在每一个随机样本上建立一个模型，最后将这些模型给汇总起来。在Bagging算法里，基于每个随机样本建立的模型往往属于同一种类型。随机森林算法是最有名的一种Bagging算法。
<br><br>
<div>
    
</div>

<strong>Boosting</strong>
<br>

在Bagging算法里，我们同时建立很多个学习模型，每个模型基于一个随机样本。而Boosting也建立很多学习模型，但这些模型是按顺序依次建立，根据前面的模型的结果，我们可以更有针对性地训练后面的模型。
<br><br>
<div>
    
</div>

<strong>Stacking</strong>
<br>

Stacking和Bagging非常相像，它也是基于随机样本建立很多个学习模型，然后将这些模型综合起来。它与Bagging不同的地方是这些模型往往不属于同一种类型。
<br><br>
<div>
    
</div>

在集成学习算法里，Bagging和Stacking是通过减少方差的方法来改善模型最终的指标，而Boosting是通过减少偏差的方法来改善模型最终的指标。
