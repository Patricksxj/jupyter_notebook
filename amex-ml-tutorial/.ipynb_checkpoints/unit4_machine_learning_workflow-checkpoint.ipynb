建立机器学习模型中至关重要的一步是如何选取数据。往往我们把所有的数据随机的分成训练样本和验证样本。模型在训练样本上进行拟合，然后在验证样本上进行验证。对于大部分建模项目，数据划分为训练样本和验证样本已经足够了，但是还有一种更为复杂，但非常有效的方法，称为交叉验证法。下面是交叉验证法的图示。

交叉验证法的关键是将数据划分成若干个部分，然后依次对其中的一部分数据进行模型拟合，然后在剩下的数据上计算模型指标。上述过程运行多遍，直到将所有部分的数据都遍历过，然后将每次运行过之后求得的模型指标汇总起来，求一个平均指标作为对模型质量的一个权衡标准。


<strong>模型验证指标</strong>

模型验证指标对于建立机器学习模型是非常关键的，指标可以用来判断我们是不是有一个好的模型，什么时候我们可以将模型训练告一段落，如果指标不好，我们应如何改善模型等等。

在业界有许多不同的模型指标，我们在这儿介绍一个非常重要的指标。


<strong>Gini 系数</strong>
<br>
Gini系数本出自于经济学的研究，它往往用来表示一个国家的财富是不是平均分布。如果一个国家的Gini系数很高，那么这个国家的贫富差距就很悬殊, 譬如最富的1%人口拥有30%的财富。如果一个国家的Gini系数很低，那这个国家的财富就非常平均分配。

Gini系数在机器学习里有相类似的含义，但区分的是模型预测值大的对象和预测值小的对象。计算给定模型Gini系数的方法是先用模型将数据样本排序，从预测值大的排到预测值小的，然后在平面上绘制一条曲线，曲线的x轴代表样本点排列的百分位，y轴代表截止到此样本点为止，包含应变量是1的百分比。Gini系数就是这条曲线和45度直线包含的面积。
<br><br>

<div>
    
</div>
<br><br>

下面用一个简单的例子说明如何计算Gini系数。假设一个样本包括10个样本点，模型对每个样本都计算一个预测值。

|应变量|模型预测值| 
|-----|--------|
|1|0.5|
|0|0.4|
|1|0.3|
|1|0.8|
|0|0.55|
|0|0.1|
|1|0.35|
|0|0.2|
|1|0.7|
|0|0.25|

用预测值对这组样本从大到小排序后变成：

|应变量|模型预测值| 
|-----|--------|
|1|0.8|
|1|0.7|
|0|0.55|
|1|0.5|
|0|0.4|
|1|0.35|
|1|0.3|
|0|0.25|
|0|0.2|
|0|0.1|

然后我们计算每一行的百分位和到这一行为止包含1的百分比

|应变量|模型预测值|百分位|包含1的百分比|
|-----|--------|-----|--------|
|1|0.8|10%|20%| 
|1|0.7|20%|40%|
|0|0.55|30%|40%|
|1|0.5|40%|60%|
|0|0.4|50%|60%|
|1|0.35|60%|80%|
|1|0.3|70%|100%|
|0|0.25|80%|100%|
|0|0.2|90%|100%|
|0|0.1|100%|100%|

通过简单的几何计算我们可以得到由最后一列构成的曲线和45度角直线所围成的面积是 0.65 - 0.5 = 0.15。

在谈到Gini系数的时候，我们往往会提到标准Gini系数。所谓标准Gini系数就是刚才计算后的结果除以Gini系数在一个样本集上可取的最大值。什么是一个样本集上最大的Gini系数呢？那就是当模型取真实值时的情况。请大家验证刚才这个例子最大Gini系数是0.25。那么它的标准Gini系数就是$\frac{0.15}{0.25} = 0.6$

Gini系数和配对比较有很强的关系。用它可以对Gini系数做一个直观的解释。我们有下面的定理：
<br><br>
$\frac{标准Gini系数+1}{2}$ = 应变量值是1的样本集中任取一个样本的模型预测值比从应变量值是0的样本集中任取一个样本的模型预测值来得大的概率。

我们用上面的例子可以来验证这个定理。在上面的例子中，取1的样本集有如下的模型预测值：
<br>
0.8，0.7，0.5，0.35，0.3

取0的样本集有如下的模型预测值：
0.55，0.4，0.25，0.2，0.1

从第一个样本集中任取一数比从第二个样本集中任取一数大的概率是多少呢？我们留给读者解答。



<strong>过拟合与欠拟合</strong>
<br>

当模型复杂度很高的时候，往往模型对训练样本拟合的很好，但一旦把这个模型运用到另一个样本集上，结果就不如人意，这称为模型过拟合。
