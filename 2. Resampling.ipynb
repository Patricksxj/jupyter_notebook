```python
import pandas as pd
import numpy as np
```

###  数据重采样 ###
- 时间数据由一个频率转换到另一个频率
- 降采样
- 升采样

```python
rng = pd.date_range('1/1/2011', periods=90, freq='D')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts.head()
```

```python
ts.resample('M').sum()
```

```python
ts.resample('3D').sum()
```

```python
day3Ts = ts.resample('3D').mean()
day3Ts
```

```python
print(day3Ts.resample('D').asfreq())
```

###  插值方法： ###
- ffill 空值取前面的值
- bfill 空值取后面的值
- interpolate 线性取值

```python
day3Ts.resample('D').ffill()
```

```python
day3Ts.resample('D').bfill(1)
```

```python
day3Ts.resample('D').interpolate('linear')
```
