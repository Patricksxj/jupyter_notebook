```python
import tushare as ts
```

```python
pabank=ts.get_hist_data('000001',start='2019-01-01',end='2019-03-11')
```

```python
pabank.describe()
```

```python
pabank.head()
```

```python
pabank.sort_index(inplace=True)
```

```python
pabank.head()
```

```python
pabank['date']=pabank.index
```

```python
pabank.describe()
```

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置matplotlib正常显示中文和负号
matplotlib.rcParams['font.sans-serif']=['Microsoft YaHei']   # 用黑体显示中文
matplotlib.rcParams['axes.unicode_minus']=False     # 正常显示负号
```

```python
fig,ax=plt.subplots(figsize=(20,15))
plt.plot(pabank['date'],pabank['close'],label='价格走势')
plt.xlabel('收盘日',fontsize=15)
plt.ylabel('股价',fontsize=18)
plt.title('19年pabank每日收盘价格变化',fontsize=20)
plt.show()
```

```python
from pyecharts import Bar, Line, Overlap
import pyecharts
line = Line("19年pabank每日收盘价_交易量变化")
bar = Bar()
line.add("股价变化",pabank['date'],pabank['close'] , is_smooth=True,mark_point=["max","min"], mark_line=[ "average"],yaxis_min=5,yaxis_max=15,is_datazoom_show=True)
bar.add("每日交易量",pabank['date'],pabank['volume'],yaxis_min=300000,yaxis_max=5000000)
overlap = Overlap()
overlap.add(line)
overlap.add(bar, yaxis_index=1, is_add_yaxis=True)
```

```python
overlap.render('2019_pabank_share_Volatility.html')
```
