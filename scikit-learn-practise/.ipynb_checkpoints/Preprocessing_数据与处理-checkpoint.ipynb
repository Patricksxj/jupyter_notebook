```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
```

```python
wine=load_wine()
```

```python
wine.feature_names
```

```python
wine.target
```

```python
df=pd.DataFrame(np.concatenate((wine.data,wine.target.reshape(-1,1)),axis=1),columns=list(wine.feature_names)+['target'])
```

```python
df.head()
```

 ###  1.minmax scaler
 - 需要2个变量，单变量不能中中心化
 - fit 只是将方法赋予了对象，并没有实现数据转化
 - fit_transform才是直接转化，没有保留原始数据

```python
scaler=MinMaxScaler()
```

```python
df[['alcohol','malic_acid']]
```

### 对函数实例化

```python
data=scaler.fit(df[['alcohol','malic_acid']])
```

### 产出两个方法，计算最大，最小值

```python
data.data_max_
```

```python
data.data_min_
```

### 对数据进行转化，不会暴力库原始数据

```python
data=data.transform(df[['alcohol','malic_acid']])
```

```python
data2=scaler.fit_transform(df[['alcohol','malic_acid']])
data2=pd.DataFrame(data2,columns=['scaler_alcohol','scaler_malic_acid'])
```

```python
data2.columns
```

### 合并新老数据集

```python
df2=pd.concat([df,data2],axis=1)
```

```python
df2.head()
```

### 方法2：standard

```python
scaler=StandardScaler()
```

```python
data3=scaler.fit_transform(df[['magnesium','total_phenols']])
data3=pd.DataFrame(data3,columns=['std_magnesium','std_total_phenols'])
```

```python
df3=pd.concat([df2,data3],axis=1)
```

```python
df3.head()
```

### 2.缺失值填充

#### 新建立一个变量，制造空数据

```python
df3['alcohol2']=df3['alcohol'].apply(lambda x:np.nan if x<=12.1 else x)
```

```python
df3.describe()
```

### 运用pandas方法填充

```python
df3['impute_alcohol2']=df3.alcohol2.fillna(df3['alcohol2'].median())
```

```python
df3.describe()
```

### 运用sklearn方法填充

```python
from sklearn.impute import SimpleImputer
```

```python
SimpleImputer?
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='median',copy=True)
impute.fit_transform(df3[['alcohol2','ash']])
```

```python
df3['sklearn_impute_alcohol2']=impute.fit_transform(df3[['alcohol2','ash']])[:,0]
```

```python
df3.describe()
```

### 3.建立分类变量

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
dummy=OneHotEncoder()
data=dummy.fit_transform(pd.concat([df3['target'],df3['target']],axis=1)).toarray()[:,:3]
```

```python
df_new=pd.DataFrame(data,columns=['target_1','target_2','target_3'])
```

```python
df3=pd.concat([df3,df_new],axis=1)
```
