```python
import pandas as pd
df = pd.DataFrame({
   "FirstName": ["Jane","john","Ashley", "MATT", "Alex"],
   "LastName": ["Doe", "Smitt", "Adams", "Tull", "Morgan"],
   "Address": ["Houston, TX", "Dallas, TX", "San Antonio, TX", 
            "Palo Alto, CA", "San Diego, CA"]})
```

```python
import numpy as np
```

```python
df1=df.copy()
df2=df.copy()
```

```python
df3=pd.concat([df1,df2])
df3
```

```python
df4=pd.concat([df1,df2],axis=1)
df4
```

```python
pd.DataFrame(np.vstack([
    df1.columns, df1, 
    df2.columns, df2]))
```

```python
pd.DataFrame(np.hstack([df1,df2]),columns=['a','b','c','d','e','f'])
```
